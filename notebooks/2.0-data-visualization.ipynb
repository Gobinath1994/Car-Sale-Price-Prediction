```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/Users/gobinathsindhuja/Desktop/Prospa_assignment/supervised_learning_project/data/vehiclesFinal.csv')
df['year'] = df['Year_Group'].astype('int64')
df.head()
```

```python
df2=df.sample(100)
df2.shape
```

```python
sns.pairplot(df2)
plt.title('Pair Plot')
plt.tight_layout()
plt.savefig('/Users/gobinathsindhuja/Desktop/Prospa_assignment/supervised_learning_project/assets/img/DataVisualization/Insight-Pair-Plot',dpi=None)
plt.show()
```

```python
sns.distplot(df.Sale_Price)
plt.tight_layout()
plt.title("Distribution of Price")
plt.savefig('/Users/gobinathsindhuja/Desktop/Prospa_assignment/supervised_learning_project/assets/img/DataVisualization/Insight-Dist-Plot',dpi=None)
plt.show()
```

```python
fig,axs=plt.subplots(nrows=2)
fig.set_size_inches(20,10)
sns.barplot(x='year',y='Sale_Price',data=df,ax=axs[0])
sns.barplot(df.groupby('year').count()['Sale_Price'].index,df.groupby('year').count()['Sale_Price'].values,ax=axs[1])
axs[0].set_title('Price vs. Year')
axs[1].set_title('Number of Cars vs. Year')
axs[1].set_ylabel('number of cars')
plt.tight_layout()
plt.savefig('/Users/gobinathsindhuja/Desktop/Prospa_assignment/supervised_learning_project/assets/img/DataVisualization/fig1.jpg',dpi=600)
plt.show()
```

```python
fig,axs=plt.subplots(nrows=2)
fig.set_size_inches(16,10)
sns.barplot(df.groupby('year').count()['Sale_Price'].index,df.groupby('year').count()['Sale_Price'].values,ax=axs[0])
sns.barplot(df.groupby('Make_Code').count()['Sale_Price'].index,df.groupby('Make_Code').count()['Sale_Price'].values,ax=axs[1])
axs[0].set_ylabel('number of cars')
axs[0].set_title('Number of Cars vs. Year')
axs[1].set_ylabel('number of cars')
axs[1].set_title('Number of Cars vs. Manufacturer')
axs[1].set_xticklabels(df.groupby('Make_Code').count()['Sale_Price'].index,rotation=90)
plt.tight_layout()
plt.savefig('/Users/gobinathsindhuja/Desktop/Prospa_assignment/supervised_learning_project/assets/img/DataVisualization/fig2.jpg',dpi=600)
plt.show()
```

```python
fig,axs=plt.subplots(nrows=2)
fig.set_size_inches(16,10)
sns.barplot(df.groupby('year').count()['Sale_Price'].index,df.groupby('year').count()['Sale_Price'].values,ax=axs[0])
sns.barplot(df.groupby('Region_Description').count()['Sale_Price'].index,df.groupby('Region_Description').count()['Sale_Price'].values,ax=axs[1])
axs[0].set_ylabel('number of cars')
axs[0].set_title('Number of Cars vs. Year')
axs[1].set_ylabel('number of cars')
axs[1].set_title('Number of Cars vs. Regional')
axs[1].set_xticklabels(df.groupby('Region_Description').count()['Sale_Price'].index,rotation=90)
plt.tight_layout()
plt.savefig('/Users/gobinathsindhuja/Desktop/A_Job_Assignment/assets/img/DataVisualization/fig3.jpg',dpi=600)
plt.show()
```
